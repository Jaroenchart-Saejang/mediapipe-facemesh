```python
import json
import base64
import os

# Load the JSON file
with open("face_capture (2).json", "r") as f:
    data = json.load(f)

# Ensure the output folder exists
output_dir = "images"
os.makedirs(output_dir, exist_ok=True)

# Loop through the sides we care about
for side in ["front", "right", "left"]:
    if side in data and "imageDataUrl" in data[side]:
        data_url = data[side]["imageDataUrl"]

        # Detect file type from prefix (png, jpeg, etc.)
        header, base64_str = data_url.split(",", 1)
        if "image/png" in header:
            ext = "png"
        elif "image/jpeg" in header or "image/jpg" in header:
            ext = "jpg"
        else:
            ext = "bin"  # fallback if unknown

        # Decode from Base64
        decoded_bytes = base64.b64decode(base64_str)

        # Save as file inside the "images" folder
        filename = os.path.join(output_dir, f"{side}.{ext}")
        with open(filename, "wb") as img_file:
            img_file.write(decoded_bytes)

        print(f"✅ {side.capitalize()} image saved as {filename}")
    else:
        print(f"⚠️ No image data found for {side}")
```
